```python
age_1 = df.loc[(df['Industry'] == 1), 'Age']
single_1 = df.loc[(df['Industry'] == 1) & (df['Marital Status'] == 1), 'Marital Status']
single_1 = single_1.rename('Single')
married_1 = df.loc[(df['Industry'] == 1) & ((df['Marital Status'] == 2) | (df['Marital Status'] == 3)), 'Marital Status']
married_1 = married_1.rename('Married')
health_1 = df.loc[(df['Industry'] == 1) & (df['Health'] > 0), 'Health']
ethnic_1 = df.loc[(df['Industry'] == 1) & (df['Ethnic Group'] > 0), 'Ethnic Group']
economic_1 = df.loc[(df['Industry'] == 1) & (df['Economic Activity'] > 0) & (df['Economic Activity'] < 4), 'Economic Activity']
occupation_1 = df.loc[(df['Industry'] == 1) & (df['Occupation'] > 0), 'Occupation']
hours_1 = df.loc[(df['Industry'] == 1) & (df['Hours worked per week'] > 0), 'Hours worked per week']
```

```python
industry_1 = pd.concat([age_1, single_1, married_1, health_1, ethnic_1, economic_1, occupation_1, hours_1], axis=1)
```

```python
industry_1.describe()
```

```python
industry = list()
for i in range(1,13):
    age = df.loc[(df['Industry'] == i), 'Age']
    single = df.loc[(df['Industry'] == i) & (df['Marital Status'] == 1), 'Marital Status']
    single = single.rename('Single')
    married = df.loc[(df['Industry'] == i) & ((df['Marital Status'] == 2) | (df['Marital Status'] == 3)), 'Marital Status']
    married = married.rename('Married')
    health = df.loc[(df['Industry'] == i) & (df['Health'] > 0), 'Health']
    ethnic = df.loc[(df['Industry'] == i) & (df['Ethnic Group'] > 0), 'Ethnic Group']
    economic = df.loc[(df['Industry'] == i) & (df['Economic Activity'] > 0) & (df['Economic Activity'] < 4), 'Economic Activity']
    occupation = df.loc[(df['Industry'] == i) & (df['Occupation'] > 0), 'Occupation']
    hours = df.loc[(df['Industry'] == i) & (df['Hours worked per week'] > 0), 'Hours worked per week']
    industry.append([age, single, married, health, ethnic, economic, occupation, hours])
```

```python
df_industry_1 = pd.DataFrame(industry[0])
df_industry_2 = pd.DataFrame(industry[1])
df_industry_3 = pd.DataFrame(industry[2])
df_industry_4 = pd.DataFrame(industry[3])
df_industry_5 = pd.DataFrame(industry[4])
df_industry_6 = pd.DataFrame(industry[5])
df_industry_7 = pd.DataFrame(industry[6])
df_industry_8 = pd.DataFrame(industry[7])
df_industry_9 = pd.DataFrame(industry[8])
df_industry_10 = pd.DataFrame(industry[9])
df_industry_11 = pd.DataFrame(industry[10])
df_industry_12 = pd.DataFrame(industry[11])
df_industry_1 = df_industry_1.transpose()
df_industry_2 = df_industry_2.transpose()
df_industry_3 = df_industry_3.transpose()
df_industry_4 = df_industry_4.transpose()
df_industry_5 = df_industry_5.transpose()
df_industry_6 = df_industry_6.transpose()
df_industry_7 = df_industry_7.transpose()
df_industry_8 = df_industry_8.transpose()
df_industry_9 = df_industry_9.transpose()
df_industry_10 = df_industry_10.transpose()
df_industry_11 = df_industry_11.transpose()
df_industry_12 = df_industry_12.transpose()
```
